```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter
```

```python
coin_df = pd.read_csv("BTCUSDT 5.csv").iloc[:, 3:]
#coin_df.resample('5T').mean().interpolate()
coin_df
```

Calculate important candle data

```python
coin_df['candle'] = np.where(coin_df['close'] > coin_df['open'], 1, 0)
coin_df['pnl'] = coin_df['close'] - coin_df['open']
coin_df['pnl %'] = coin_df['pnl']/coin_df['open']

print(f"99,9% of PnL% is between {round(np.percentile(coin_df['pnl %'], 0.05), 5)} and \
    {round(np.percentile(coin_df['pnl %'], 99.95), 5)}")

print(f"99% of PnL% is between {round(np.percentile(coin_df['pnl %'], 0.5), 5)} and \
    {round(np.percentile(coin_df['pnl %'], 99.5), 5)}")

print(f"90% of PnL% is between {round(np.percentile(coin_df['pnl %'], 10), 5)} and \
    {round(np.percentile(coin_df['pnl %'], 95), 5)}")

print(f"80% of PnL% is between {round(np.percentile(coin_df['pnl %'], 5), 5)} and \
    {round(np.percentile(coin_df['pnl %'], 90), 5)}")

plt.hist(coin_df['pnl %'], bins = 40, range = (-0.01, 0.01), rwidth = 0.8,
         weights=np.ones_like(coin_df['pnl %']) / len(coin_df['pnl %']))
plt.show()
```

```python
from ta.trend import SMAIndicator
from ta.volatility import AverageTrueRange

coin_df['SMA 12'] = SMAIndicator(coin_df['close'], 12).sma_indicator()
coin_df['SMA 36'] = SMAIndicator(coin_df['close'], 36).sma_indicator()

coin_df['ATR 144'] = AverageTrueRange(coin_df['high'], coin_df['low'],
                                      coin_df['close'], 144).average_true_range().replace(0, np.NaN)

coin_df['C - SMA 12'] = (coin_df['close'] - coin_df['SMA 12'])
coin_df['(C - SMA 12)/ATR 144'] = (coin_df['C - SMA 12']/coin_df['ATR 144'])
coin_df['C - SMA 36'] = (coin_df['close'] - coin_df['SMA 36'])
coin_df['(C - SMA 36)/ATR 144'] = (coin_df['C - SMA 36']/coin_df['ATR 144'])

coin_df['candle'] = np.where(coin_df['close'] > coin_df['open'], 1, 0)

plt.plot(coin_df['close'].iloc[-500:])
plt.plot(coin_df['SMA 12'].iloc[-500:])
plt.plot(coin_df['SMA 36'].iloc[-500:])
```

```python
analysis_df = coin_df[['(C - SMA 12)/ATR 144', '(C - SMA 36)/ATR 144']].shift(1)
analysis_df['candle'] = coin_df['candle']

analysis_df.iloc[140:155]
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
     analysis_df[['(C - SMA 12)/ATR 144', '(C - SMA 36)/ATR 144']].iloc[200:],
     analysis_df['candle'].iloc[200:], test_size=0.1, shuffle = False)

forest_clf = RandomForestClassifier(n_estimators = 300, max_leaf_nodes = 2, max_depth = 2, min_samples_leaf = 100, random_state = 1)
lr_clf = LogisticRegression(solver ='lbfgs', multi_class ='multinomial', max_iter = 200, random_state = 1)
#svc_clf = SVC(gamma ='auto', probability = True, random_state = 1)

estimators = [('lr_clf', lr_clf), ('forest_clf', forest_clf)]
```

```python
vot_hard = VotingClassifier(estimators = estimators, voting ='soft', verbose = True)
vot_hard.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = vot_hard.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
backtest_df = coin_df.iloc[-X_test.shape[0]:, 0:8]

backtest_df['candle_pred'] = y_pred
#X_test['candle'] = y_test
backtest_df
```

```python
pnl = np.where(backtest_df['candle'] == backtest_df['candle_pred'],
                np.absolute(backtest_df['close'] - backtest_df['open']),
               -np.absolute(backtest_df['close'] - backtest_df['open']))
backtest_df['strategy pnl'] = pnl
backtest_df['strategy pnl %'] = backtest_df['strategy pnl']/backtest_df['open']
backtest_df
```

```python
start_cap = 1000
invest_perc = 0.4
strategy_backtest = [start_cap]

for i in range(backtest_df.shape[0]):
    cap = strategy_backtest[-1]
    if backtest_df['strategy pnl %'].iloc[i] < -0.01:
        strategy_backtest.append(cap*(1 + invest_perc*(0.99 - 1)))
        continue
    else:
        strategy_backtest.append(cap*(1 + invest_perc*((1 + backtest_df['strategy pnl %'].iloc[i]) - 1)))

plt.plot(strategy_backtest)
plt.axhline(y = start_cap, color = 'r', linestyle = '--')
```
